```python
#pair plot

#if there are many variables\,it is not convinient to draw scatter plots for each pair of variables to understand the relationship ,so a pair plot can be use to depict the relationship in a single dig which can be plotted using pairplot() method
```

```python
import seaborn as sn
influential_features=['SR_B','AVE','SIXERS','SOLD PRICE']
sn.pairplot(ipl_auction_df[influential_features])
plt.show()
```
