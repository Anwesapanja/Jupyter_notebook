```python
#Draw two plots using subplot() function.
import matplotlib.pyplot as plt
import numpy as np
#plot1:
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])

plt.subplot(1,2,2)
plt.plot(x,y)
plt.title("INCOME")

plt.show()
```

```python
#add a title for the entire fig

x=np.array([0,1,2,3])
y=np.array([3,8,1,10])

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])

plt.subplot(1,2,2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("Pattern of Sales-Income")
plt.show()


```

```python
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])

plt.subplot(2,1,1)
plt.plot(x,y)
plt.title("Sales Data")

plt.subplots_adjust(hspace=0.5)

#plot 2
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])

plt.subplot(2,1,2)
plt.plot(x,y)
plt.title("INCOME Data")

plt.suptitle("Pattern of Sales-Income")
plt.show()
```
