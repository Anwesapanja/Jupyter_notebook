```python
import seaborn as sn
```

```python
sn.barplot(x='AGE', y='SOLD PRICE', data=soldprice_by_age)
plt.title("bar plot for aversge sold price versus age")
plt.xlabel('AGE')
plt.ylabel('mean(SOLD PRICE)')
plt.show()
```

```python
# How to draw histogram 
plt.hist(ipl_auction_df['SOLD PRICE'])
plt.title("Histogram for sold price")
plt.xlabel('SOLD PRICE')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(ipl_auction_df['SOLD PRICE'],bins=20)
plt.title("Histogram for sold price")
plt.xlabel('SOLD PRICE')
plt.ylabel('Frequency')
plt.show()
```

```python
# distribution or density plot
sn.distplot(ipl_auction_df['SOLD PRICE'])
plt.title("Distribution plot for sold price")
plt.show()
```

```python
# scattered plot: in a scatterd plot the values of two variables are plotted along two axes & resulting pattern can reveal correlation between the variables. It is also useful assessing the strength of the relationship and to find if there any outliers into the dataset\

x= np.array([5,2,6,1,2,3,7,2,4])
y=np.array([4,5,2,3,7,5,4,6,7])
plt.scatter(x,y)
plt.show
```
